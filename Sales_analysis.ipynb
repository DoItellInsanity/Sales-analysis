```python
import pandas as pd
import os
```

```python
folder_path = r'C:\Users\User\OneDrive\Desktop\ML\learning\Pandas\Pandas-Data-Science-Tasks-master\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data'

# List all files in the folder
folders = [file for file in os.listdir(folder_path)]

# Initialize an empty DataFrame
yearly_data = pd.DataFrame()

# Loop through each file in the folder and concatenate the data
for file in folders:
    file_path = os.path.join(folder_path, file)  # Concatenate folder path and file name
    data = pd.read_csv(file_path)
    yearly_data = pd.concat([yearly_data, data])

# Save the concatenated data to a CSV file
yearly_data.to_csv('yearly_data.csv', index=False)
```

```python
#Clean Data
nan_data = yearly_data[yearly_data.isna()]
nan_data
```

```python
clean_na = yearly_data.dropna()
clean_na
```

```python
# Searching and Removing 'Or' from the data
clean_na =  clean_na[clean_na['Order Date'].str[0:2]!= 'Or']
clean_na.head()
```

```python
clean_na.head()
```

```python
# Add month data with an additional column
clean_na['Month'] = clean_na['Order Date'].str[0:2]
clean_na['Month'] = clean_na['Month'].astype('int64')
clean_na.tail()
```

```python
clean_na.info()
```

```python
# Convert Quantity Ordered and Price Each in correct datatype
clean_na['Quantity Ordered'] = pd.to_numeric(clean_na['Quantity Ordered'])
clean_na['Price Each'] = pd.to_numeric(clean_na['Price Each'])
clean_na.head()
```

```python
clean_na.info()
```

Sales Per order

```python
clean_na['sales per order']  = clean_na['Quantity Ordered'] * clean_na['Price Each']
clean_na.head()
```

Best Month for Sales

```python
sales_per_month=clean_na.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months, sales_per_month['sales per order'])
plt.xticks(months)
plt.ylabel('Sales in millions(USD)')
plt.show()
```

CIty with highest number of sales?

```python
# Add city column 
clean_na['City'] = clean_na['Purchase Address'].apply(lambda x : x.split(',')[1] )
clean_na.head()
```

```python
sales_per_city = clean_na.groupby('City').sum()['sales per order']
sales_per_city.head(10)
```

```python
import matplotlib.pyplot as plt
city= sales_per_city['City'].unique()
plt.bar(city, sales_per_month['sales per order'])
plt.xticks(city,rotation='vertical')
plt.xlabel('City Name')
plt.ylabel('Sales in millions(USD)')
plt.show()
```

Best time to show Ad 

```python
clean_na['Order Date'] = pd.to_datetime(clean_na['Order Date'])
```

```python
clean_na['Hour']= clean_na['Order Date'].dt.hour
clean_na['Minute']= clean_na['Order Date'].dt.minute
clean_na.head()
```

```python
hours = [Hour for Hour, df in clean_na.groupby('Hour')]
plt.plot(hours,clean_na.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Sales Count')
plt.grid()
plt.show()
```
